```python
from fastai import *
from fastai.vision import *
from fastai.metrics import error_rate
import os
import pandas as pd
import numpy as np
```

```python
x  = '/kaggle/input/cat-and-dog/training_set/training_set'
path = Path(x)
path.ls()
```

```python
np.random.seed(40)
data = ImageDataBunch.from_folder(path, train = '.', valid_pct=0.2,
                                  ds_tfms=get_transforms(), size=224,
                                  num_workers=4).normalize(imagenet_stats)
```

```python
data.show_batch(rows=3, figsize=(7,6),recompute_scale_factor=True)
```

```python
learn = cnn_learner(data, models.resnet18, metrics=[accuracy], model_dir = Path('../kaggle/working'),path = Path("."))
```

```python
learn
```

```python
learn.lr_find()
learn.recorder.plot(suggestions=True)
```

```python
lr1 = 1e-4
lr2 = 1e-1
learn.fit_one_cycle(4,slice(lr1,lr2))
```

```python
img = open_image('/kaggle/input/cat-and-dog/training_set/training_set/dogs/dog.3235.jpg')
print(learn.predict(img)[0])
img
```
